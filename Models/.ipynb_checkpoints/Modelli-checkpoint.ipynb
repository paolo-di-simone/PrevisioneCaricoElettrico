### Definizione funzioni d'utilità

```python
def save_scaler(scaler, path):
    with open(path, "wb") as file:
        pickle.dump((scaler.data_min_, scaler.data_max_), file)
```

```python
def read_scaler(path):
    with open(path, "rb") as file:
        data_min, data_max = pickle.load(file)
        scaler = MinMaxScaler()
        scaler.data_min_ = data_min
        scaler.data_max_ = data_max
        return scaler
```

```python
def plot_distribution(df, fields, start_point, end_point, dim, title, xlabel, ylabel):
    
    df_tmp = df.loc[start_point:end_point]
    
    plt.figure(figsize=dim) 
    plt.grid()
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    for field in fields:
        plt.plot(df_tmp.index, df_tmp[field], linewidth="2", label=field)
    
    plt.xticks(rotation=45)
    
    plt.legend()
    
    #file_name = "".join(title.lower()).replace(" ", "_")
    #path = make_path(plot_path, file_name)
    #plt.savefig(path, bbox_inches='tight', transparent=True)
    plt.show()
```

```python
def plot_predict(values_true, values_predict, c1, c2, title, xlabel, ylabel, dim):
                
    plt.figure(figsize=dim) 
    plt.grid(color=c2, linestyle="-", linewidth="0.2", axis="y")
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    len_x = len(values_true)
    xvalue = [x for x in range(0,len_x)]  

    plt.plot(xvalue, values_true, color=c1, linewidth="2")
    plt.plot(xvalue, values_predict, color="red", linewidth="2")
    
    plt.legend(["Carico", "Predizione"], loc="upper right")
    
    #file_name = "".join(title.lower()).replace(" ", "_")
    #path = make_path(plot_path, file_name)
    #plt.savefig(path, bbox_inches='tight', transparent=True)
    plt.show()
```

```python
def plot_history(h, c1, c2):
    # summarize history for mape
    plt.figure(figsize=(10,5))
    plt.grid()
    plt.plot(h["mape"], color=c1, linewidth="2", marker='o')
    plt.plot(h["val_mape"], color=c2, linewidth="2", marker='o')
    plt.title("Model MAPE")
    plt.ylabel("MAPE(%)")
    plt.xlabel("epochs")
    plt.legend(["train", "validation"], loc="upper right")
    plt.show()
    # summarize history for loss
    plt.figure(figsize=(10,5))
    plt.grid()
    plt.plot(h["loss"], color=c1, linewidth="2", marker='o')
    plt.plot(h["val_loss"], color=c2, linewidth="2", marker='o')
    plt.title("Model Loss")
    plt.ylabel("loss")
    plt.xlabel("epoch")
    plt.legend(["train", "validation"], loc="upper right")
    plt.show()
```

```python
def add_prediction(df, prediction, time_steps, label):
    df = df.iloc[time_steps:]
    df[label] = prediction
    return df
```

### Import dei dati

```python
dataset_path = os.path.join("..", "Preprocessing", "Dataset", "dataset_training.csv")
df = pd.read_csv(dataset_path, index_col="Time")
df.index = pd.to_datetime(df.index)
df = df.loc[df.index.year >= 2021]
df
```

### OneHot encoding e normalizzazione

```python
encoder = OneHotEncoder()
df["HourEncoded"] = encoder.fit_transform(np.array(df["Hour"]).reshape(-1,1)).toarray().tolist()
df["MonthEncoded"] = encoder.fit_transform(np.array(df["Month"]).reshape(-1,1)).toarray().tolist()
df["DayOfWeekEncoded"] = encoder.fit_transform(np.array(df["DayOfWeek"]).reshape(-1,1)).toarray().tolist()
```

```python
# Ricordare di fare la fit su tutti i dati (e non solo train o test)
scaler1 = MinMaxScaler()
df["L1Scaled"] = scaler1.fit_transform(df[["L1"]])
scaler2 = MinMaxScaler()
df["L2Scaled"] = scaler2.fit_transform(df[["L2"]])
scaler3 = MinMaxScaler()
df["L3Scaled"] = scaler3.fit_transform(df[["L3"]])
```

```python
df
```

### Splitting in training, validation e test

```python
train_ration = 0.8
validation_ratio = 0.1
test_ratio = 0.1
```

```python
train_size = int(len(df) * train_ration)
validation_size = int(len(df) * validation_ratio)
test_size = int(len(df) * test_ratio)
df_train = df.iloc[0:train_size]
df_validation = df.iloc[train_size:train_size+validation_size]
df_test = df.iloc[train_size+validation_size:]
print("Train size:", len(df_train), "\nValidation size:", len(df_validation), "\nTest size:", len(df_test))
```

### Creazione finestre e reshaping

```python
def windowed_dataset(df, time_steps):
    y_L1 = []
    y_L2 = []
    y_L3 = []
    x_time_series_L1 = []
    x_time_series_L2 = []
    x_time_series_L3 = []
    x_other_feature = []
    for i in range(len(df)-time_steps):
        
        x_time_series_L1.append(df["L1Scaled"].iloc[i:i+time_steps])
        x_time_series_L2.append(df["L2Scaled"].iloc[i:i+time_steps])
        x_time_series_L3.append(df["L3Scaled"].iloc[i:i+time_steps])
        
        tmp = []
        tmp.extend(df["HourEncoded"].iloc[i+time_steps])
        tmp.extend(df["DayOfWeekEncoded"].iloc[i+time_steps])
        tmp.extend(df["MonthEncoded"].iloc[i+time_steps])
        x_other_feature.append(tmp)
        
        y_L1.append(df["L1Scaled"].iloc[i+time_steps])
        y_L2.append(df["L2Scaled"].iloc[i+time_steps])
        y_L3.append(df["L3Scaled"].iloc[i+time_steps])
    
    return {
        "xL1": np.array(x_time_series_L1),
        "xL2": np.array(x_time_series_L2),
        "xL3": np.array(x_time_series_L3),
        "OtherFeature": np.array(x_other_feature),
        "yL1": np.array(y_L1),
        "yL2": np.array(y_L2),
        "yL3": np.array(y_L3)
    }
```

```python
def get_x(data, line):
    return [data["x"+line], data["OtherFeature"]]
def get_y(data, line):
    return data["y"+line]
```

```python
time_steps = 7 * 24
dim_other_feature = 24 + 7 + 12
```

```python
train = windowed_dataset(df_train, time_steps)
validation = windowed_dataset(df_validation, time_steps)
test = windowed_dataset(df_test, time_steps)
```

### Test 1

```python
input_load = Input(shape=(time_steps,1), name="input_load")

lstm1 = LSTM(units=time_steps, return_sequences=True)(input_load)
lstm2 = LSTM(units=time_steps)(lstm1)
dropout = Dropout(rate=0.2)(lstm2)

dense = Dense(units=1, activation="linear")(dropout)

model = Model(inputs=input_load, outputs=dense)
```

```python
model.summary()
```

```python
model_image_path = os.path.join(model_image_base_path, "test1.png")
plot_model(model, to_file=model_image_path, show_shapes=True)
```

```python
model.compile(loss="mse", optimizer="adam", metrics=["mape"])
```

```python
history = model.fit(
    get_x(train, "L1")[0], get_y(train, "L1"), 
    epochs=5, 
    verbose=1, 
    batch_size=128, 
    validation_data=(get_x(validation, "L1")[0], get_y(validation, "L1")), 
    shuffle=False
)
```

```python
plot_history(history.history, "#0F52BA", "#50C878")
```

```python
model.evaluate(get_x(test, "L1"), get_y(test, "L1"))
```

```python
y_pred_test = model.predict(get_x(test, "L1"))
y_pred_test_inv = scaler1.inverse_transform(y_pred_test)
```

```python
df_prediction = add_prediction(df_test, y_pred_test_inv, time_steps, "PowerL1Predicted")
```

```python
title = "Curva di carico giornaliera"
xlabel = "TotalPower"
ylabel = "W"
dim = (15,5)
c1 = "#2271b3"
c2 = "#000000"
plot_predict(df_prediction["PowerL1"], df_prediction["PowerL1Predicted"], c1, c2, title, xlabel, ylabel, dim)
```

### Test 2

```python
# Pipeline LOAD
input_load = Input(shape=(time_steps, 1), name="input_load")
c1 = Conv1D(32, kernel_size=3, activation="relu")(input_load)
m1 = MaxPooling1D(pool_size=2)(c1)
c2 = Conv1D(64, kernel_size=3, activation="relu")(m1)
m2 = MaxPooling1D(pool_size=2)(c2)
c3 = Conv1D(128, kernel_size=3, activation="relu")(m2)
flatten1 = Flatten(name="flatten1")(c3)
dropout = Dropout(rate=0.2)(flatten1)

# Pipeline OTHER FEATURE
input_other_feature = Input(shape=(dim_other_feature, ), name="input_other_feature")
dense1 = Dense(64, activation="relu", name="dense1")(input_other_feature)
dense2 = Dense(32, activation="relu", name="dense2")(dense1)
flatten2 = Flatten(name="flatten2")(dense2)

# Concatenate pipelines
concatenated_layer = concatenate([dropout,flatten2], axis=1, name="concatenated_layer")

dense3 = Dense(units=1, activation="linear", name="dense3")(concatenated_layer)

model = Model(inputs=[input_load, input_other_feature], outputs=dense3)
```

```python
model.summary()
```

```python
model_image_path = os.path.join(model_image_base_path, "test2.png")
plot_model(model, to_file=model_image_path, show_shapes=True)
```

```python
def root_mean_squared_error(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true))) 
```

```python
model.compile(loss=root_mean_squared_error, optimizer="adam", metrics=["mape"])
```

```python
history = model.fit(
    get_x(train, "L3"), get_y(train, "L3"), 
    epochs=100, 
    verbose=1, 
    batch_size=128, 
    validation_data=(get_x(validation, "L3"), get_y(validation, "L3")), 
    shuffle=False
)
```

```python
plot_history(history.history, "#0F52BA", "#50C878")
```

```python
model.evaluate(get_x(test, "L3"), get_y(test, "L3"))
```

### PAPER13

```python
# PIPELINE LOAD
input_load = Input(shape=(time_steps, 1), name="input_load")

# Pipeline 1
branch1_1 = Conv1D(filters=64, kernel_size=3, activation='relu', name="branch1_1")(input_load)
branch1_2 = MaxPooling1D(pool_size=2, strides=2, name="branch1_2")(branch1_1)

# Pipeline 2
branch2_1 = Conv1D(filters=64, kernel_size=5, activation='relu', name="branch2_1")(input_load)
branch2_2 = MaxPooling1D(pool_size=2, strides=2, name="branch2_2")(branch2_1)

# Pipeline 3
branch3_1 = Conv1D(filters=64, kernel_size=7, activation='relu', name="branch3_1")(input_load)
branch3_2 = MaxPooling1D(pool_size=2, strides=2, name="branch3_2")(branch3_1)

# Concatenate pipelines
concatenated_layer1 = concatenate([branch1_2,branch2_2,branch3_2], axis=1, name="concatenated_layer1")

lstm_layer = LSTM(64, return_sequences=True, name="lstm_layer")(concatenated_layer1)

global_average_pooling_layer = GlobalAveragePooling1D(name="global_average_pooling_layer")(lstm_layer)
```

```python
# PIPELINE OTHER FEATURE
input_other_feature = Input(shape=(dim_other_feature, ), name="input_other_feature")
dense1 = Dense(64, activation="relu", name="dense1")(input_other_feature)
flatten1 = Flatten(name="flatten1")(dense1)
```

```python
# CONCATENATE PIPLINES
concatenated_layer2 = concatenate([global_average_pooling_layer,flatten1], axis=1, name="concatenated_layer2")

dense2 = Dense(64, activation="relu", name="dense2")(concatenated_layer2)
dense3 = Dense(64, activation="relu", name="dense3")(dense2)
output_layer = Dense(1, activation="linear", name="output_layer")(dense3)

model = Model(inputs=[input_load, input_other_feature], outputs=[output_layer])
```
